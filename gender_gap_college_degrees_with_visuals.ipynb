```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
# all_cats = stem_cats + lib_arts_cats + other_cats

fig = plt.figure(figsize=(16, 16))

stem_val = 0
lib_val = 0
other_val = 0
for i in range(0,18):
    if i in [0,3,6,9,12,15]:
        ax = fig.add_subplot(6,3,i+1)
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[stem_val]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[stem_val]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(stem_cats[stem_val])
        stem_val +=1
    if i in [1,4,7,10,13]:
        ax = fig.add_subplot(6,3,i+1)
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[lib_val]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[lib_val]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(lib_arts_cats[lib_val])
        lib_val +=1
    if i in [2,5,8,11,14,17]:
        ax = fig.add_subplot(6,3,i+1)
        ax.plot(women_degrees['Year'], women_degrees[other_cats[other_val]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[other_val]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(other_cats[other_val])
        other_val +=1
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
    
    if i == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    if i == 1:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    if i== 2:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    if i == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        ax.tick_params(labelbottom="on")
    if i == 17:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        ax.tick_params(labelbottom="on")
    if i == 13:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        ax.tick_params(labelbottom="on")
plt.show()
```
